### Batch Running of Experiments

```python
# parameters: name, iterations, agent_range, prey_reward, max_age, alpha, beta, gamma
```

```python
batch_start_time = time.time()
alphas = [.5] #0, .25, .5, .75
irs = [75, 100] #25, 50, 
for alpha in alphas:
  for ir in irs:
    experiment_start_time = time.time()
    name = f'exp_a{int(alpha*100):03}_ir{ir:03}'
    cmd = f'python schelling_example.py {name} 5000 5 {ir} 100 {alpha} 1 1'
    print(cmd)
    os.system(cmd)
    seconds = round(time.time() - experiment_start_time)
    print(f"   Experiment duration: {seconds/60:.2f} minutes")    

seconds = round(time.time() - batch_start_time)
print(f'Finished {len(alphas)*len(irs)} experiments.')
print(f"Batch Experiment duration: {seconds/60:.2f} minutes")        
```
