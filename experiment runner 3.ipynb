## Run by alpha and IR value lists

```python
batch_start_time = time.time()
alphas = [1] 
irs = [25, 50, 75, 100] 
iters = 5000
agent_range = 5
max_age = 100 #.default=100
p_hunter = 0.05 #.default=0.05
p_resurrection = 1 #.default=0.2
sr_multiplier = 1 #.default=1
print(f'Running {len(alphas)*len(irs)} experiments...')
for alpha in alphas:
  for ir in irs:
    experiment_start_time = time.time()
    name = f'exp_a{int(alpha*100):03}_ir{ir:03}'
    name += f'_ag{max_age:03}'
    name += f'_ph{int(p_hunter*100):03}'
    name += f'_pr{int(p_resurrection*100):03}'
    name += f'_sx{int(sr_multiplier*100):03}'
    cmd = f'python schelling_example.py {name} {iters} {agent_range} {ir} {max_age} {alpha} 1 1'
    cmd += f' {p_hunter} {p_resurrection} {sr_multiplier}'
    curr_time = time.strftime("%H:%M:%S", time.localtime())
    print(f'{cmd}  (started {curr_time})')
    os.system(cmd)
    seconds = round(time.time() - experiment_start_time)
    print(f"   Experiment duration: {seconds/60:.2f} minutes")    

seconds = round(time.time() - batch_start_time)
print(f'Finished {len(alphas)*len(irs)} experiments.')
print(f"Batch Experiment duration: {seconds/60:.2f} minutes")        
```

## Run by Parameter Tuples

```python
batch_start_time = time.time()
iters = 5000
agent_range = 5
max_age = 100 #.default=100
p_hunter = 0.1 #.default=0.05
p_resurrection = 0.2 #.default=0.2
sr_multiplier = 1 #.default=1
parameters = ((.50, 50),    
              (.75,0),
              (.75,50),
              (1.00,50),
              (1.00,75),
              ) #add comma if there is only one element
print(f'Running {len(parameters)} experiments...')
for params in parameters:
  alpha, ir = params
  experiment_start_time = time.time()
  name = f'exp_a{int(alpha*100):03}_ir{ir:03}'
  name += f'_ag{max_age:03}'
  name += f'_ph{int(p_hunter*100):03}'
  name += f'_pr{int(p_resurrection*100):03}'
  name += f'_sx{int(sr_multiplier*100):03}'
  cmd = f'python schelling_example.py {name} {iters} {agent_range} {ir} {max_age} {alpha} 1 1'
  cmd += f' {p_hunter} {p_resurrection} {sr_multiplier}'
  curr_time = time.strftime("%H:%M:%S", time.localtime())
  print(f'{cmd}  (started {curr_time})')
  os.system(cmd)
  seconds = round(time.time() - experiment_start_time)
  print(f"   Experiment duration: {seconds/60:.2f} minutes")    

seconds = round(time.time() - batch_start_time)
print(f'Finished {len(parameters)} experiments.')
print(f"Batch Experiment duration: {seconds/60:.2f} minutes") 
```
