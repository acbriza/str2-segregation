The crystal is a four dimensional tensor of the simulation where the dimensions are as following: (ITERATION_INDEX, HORIZONTAL_LOCATION, VERTICAL_LOCATION, CHANNEL). 

CHANNEL = {0: agent type, 1: agent age, 2: agent id}. 

For example the slice slice = crystal[0, 5, 5, :] gives the type, age and id of the agent at location (5,5) in the first iteration. If the location is not occupied by an agent, then the slice is a zero

```python
f = gzip.open('my_experiment/crystal.npy.gz','rb')
crystal = np.load(f)
type(crystal), crystal.shape  #iterations, env_x, env_y, 
```

```python
# crystal[0, 5, 5, :] gives the type, age and id of the agent at location (5,5) in the first iteration
crystal[5, 5, 5, :]
```

```python
for i in range(50):
  c = crystal[5, i, 5, :]
  if c.sum() !=0:
    print(i)
    break
```
