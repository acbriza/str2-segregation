## Run by alpha and IR value lists

```python
# parameters: name, iterations, agent_range, prey_reward, max_age, alpha, beta, gammab
```

```python
batch_start_time = time.time()
iters = 5000 #default
#iters = 10 #simulation
alphas = [1]
irs = [0, 25, 50, 75, 100]
teleport = 'fov'
print(f'Running {len(alphas)*len(irs)} experiments...')
for alpha in alphas:
  for ir in irs:
    experiment_start_time = time.time()
    name = 'exp' if iters==5000 else 'sim'    
    name += f'_a{int(alpha*100):03}_ir{ir:03}'
    name += f'__nh_{teleport}'
      
    cmd = f'python schelling_example.py {name} {iters} 5 {ir} 100 {alpha} 1 1'
    curr_time = time.strftime("%H:%M:%S", time.localtime())
    print(f'{cmd}  (started {curr_time})')
    os.system(cmd)
    seconds = round(time.time() - experiment_start_time)
    print(f"   Experiment duration: {seconds/60:.2f} minutes")    

seconds = round(time.time() - batch_start_time)
print(f'Finished {len(alphas)*len(irs)} experiments.')
print(f"Batch Experiment duration: {seconds/60:.2f} minutes")        
```
